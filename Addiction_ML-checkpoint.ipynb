𝐑𝐞𝐠𝐫𝐞𝐬𝐬𝐢𝐨𝐧 𝐓𝐚𝐬𝐤𝐬(𝐌𝐨𝐝𝐞𝐥𝐬 𝐛𝐮𝐢𝐥𝐝𝐢𝐧𝐠)

```python
df2= df1.copy()
df2.head()
```

```python
education_map = {
    'Primary': 1,
    'High School': 2,
    'Secondary': 3,
    'College': 4,
    'University': 5,
    'Postgraduate': 6,
    'Unknown': 0  # or maybe np.nan, if you want to drop/impute
}

exercise_map = {
    'Never': 1,
    'Rarely': 2,
    'Weekly': 3,
    'Daily': 4
}
diet_map = {'Poor': 1, 'Average': 2, 'Good': 3}
mental_map = {'Poor': 1, 'Average': 2, 'Good': 3}

employment_map= {'Employed': 3,
   'Retired': 4,
   'Self-Employed': 2,
   'Student': 1,
   'Unemployed': 0}
marital_map= {'Divorced': 3,
   'In a relationship': 1,
   'Married': 2,
   'Single': 0,
   'Widowed': 4}
has_health_issues_map= {False: 0, True: 1}
social_support_map= {'Moderate': 2, 'None': 0, 'Strong': 3, 'Weak': 1}
therapy_history_map= {'Current': 1, 'None': 0, 'Past': 2}

df2['mental_health_encoded'] = df2['mental_health_status'].map(mental_map)
df2['diet_quality_encoded'] = df2['diet_quality'].map(diet_map)
df2['education_encoded'] = df2['education_level'].map(education_map)
df2['employment_encoded'] = df2['employment_status'].map(employment_map)
df2['marital_encoded'] = df2['marital_status'].map(marital_map)
df2['health_encoded'] = df2['has_health_issues'].map(has_health_issues_map)
df2['social_encoded'] = df2['social_support'].map(social_support_map)
df2['therapy_encoded'] = df2['therapy_history'].map(therapy_history_map)



```

```python
df3 = df2.copy()
df3.head(2)
```

```python
from sklearn.ensemble import RandomForestRegressor
import matplotlib.pyplot as plt

X = df3[['age', 'bmi', 'sleep_hours',
         'drinks_per_week', 'children_count',
         'mental_health_encoded', 'diet_quality_encoded', 'education_encoded',
         'employment_encoded', 'marital_encoded', 'health_encoded',
         'social_encoded', 'therapy_encoded']]
y = df3['smokes_per_day']

model = RandomForestRegressor(random_state=42)
model.fit(X, y)

importances = model.feature_importances_
features = X.columns

for feat, score in zip(features, importances):
    print(f'{feat}: {score:.3f}')

plt.figure(figsize=(10, 6))
plt.barh(features, importances, color='skyblue')
plt.xlabel("Feature Importance")
plt.title("Random Forest - Feature Importance for 'smokes_per_day'")
plt.gca().invert_yaxis()
plt.tight_layout()
plt.show()
```
